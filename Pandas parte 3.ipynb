## Metodos Unique & Nunique

**"unique()" nos permite mostrar en una lista, los valores unicos que aparecen en una columna (llamese una lista de los posibles valores que puede adquirir esa columna en los registros segun la info en el dataframe)** <br><br>

***Por ejemplo:***

```python
airbnb['Neighbourhood'].unique()
```

**Por otro lado esta "nunique()", que nos permite contar cuantos valores unicos hay**<br><br>

***Por ejemplo:***

```python
airbnb['Neighbourhood'].nunique()
```

***IMPORTANTE: unique() considera los valores NaN, por lo que de no limpiar el dataset antes, estos seran conciderados en la lista, como podemos ver abajo***

```python
airbnb['Beds'].unique()
```

**IMPORTANTE: nunique(), por defecto, no cuenta los NaN, si los queremos considerar en el conteo, debemos especificarlo pasandole el parametro dropna, que utiliza un bool para activar o desactivar esta funcion** <br><br>

***Ejermplos para ambos casos:***

```python
airbnb['Beds'].nunique() #No cuenta los NaN
```

```python
airbnb['Beds'].nunique(dropna=False) #Cuenta los NaN, ya que estamos especificando que desactive la eliminacion de los mismos
```

## Filtrar data con metodo Between

**El metodo Between funciona similar a isna(), respondiendonos con un boolean si el valor en la columna cumple o no con el filtro que estamos dandole, en este caso between, que significa "entre"** <br><br>

***Por ejemplo: consultemos filtrando por precio del airbnb, para esto, usamos el metopdo between(), especificando dentro del parentesis entre que valores estaran los valores de precio de cada registro***

```python
airbnb['Price'].between(100, 200) #Como se puede ver aqui estoy consultando si el valor de la columna price del registro esta entre 100 y 200
```

<br>*Se ve algo feo, pero lo podemos arreglar con otro set de airbnb + corchetes* <br>

```python
airbnb[airbnb['Price'].between(100, 200)] #Con este set extra de corchetes y nombre del dataframe veremos los resultados mas ordenados
```

**Como se puede comprobar arriba, esta vez, el resultado nos trajo solo los registros con un precio de entre 100 y 200 USD**<br><br>

**A este metodo podemos agregar el metodo value_counts(), de modo que podamos contar cuantos registros hay de cada valor unico en la columna**<br><br>

***Por ejemplo***

```python
airbnb[airbnb['Price'].between(100, 200)]['Price'].value_counts()
```

## Utilizando multiples filtros

Para utilizar varios filtro podemos usar el simbolo and "&"

```python
airbnb[airbnb['Price'].between(100, 200) & airbnb['Beds'].between(1,2)]
```

**Ademas de usar Between, podemos tambien usar valores iguales con "==".**<br><br>

***Por ejemplo:***

```python
airbnb[airbnb['Price'].between(100, 200) & airbnb['Beds'] == 1]
```

***Otro ejemplo:***

```python
airbnb[(airbnb['Property Type'] == 'Apartment') & (airbnb['Neighbourhood'] == 'Manhattan')]
```

**IMPORTANTE:**<BR><BR>
En Python:

() → paréntesis → agrupan expresiones

[] → indexación (acceder a algo)

& → operador lógico AND (bit a bit)


Los corchetes NO sirven para agrupar condiciones, solo para indexar.
<br><br>
Por eso los filtros o condiciones van en parentesis simple y no corchete o llaves

**Podemos agregar cuantas condiciones queramos usando &**

## Filtros avanzados con Isin y Or (representado por el simpbolo "|")

```python
salaries
```

***Digamos que deseo filtrar de este dataframe solo los registros que tenga como titulo universitario Data engineer y Data Scientist, para esto podemos usar or***<br><br>

***Por ejemplo:***

```python
salaries[(salaries['job_title'] == 'Data Scientist') | (salaries['job_title'] == 'Data Engineer')]
```

**Incluso podemos meterle un and a alguna de las condiciones** <br><br>

***Por ejemplo:***

```python
salaries[((salaries['job_title'] == 'Data Scientist') & (salaries['salary_in_usd'] == 212000)) | 
    ((salaries['job_title'] == 'Data Engineer') & (salaries['salary_in_usd'] == 300000))] 
#Estas dos lineas de codigo se pueden poner en una, pero se ve mas ordenado en dos a mi gusto
```

**De querer contar los resultados que hacen match, tambien se puede hacer, mediante value_counts()**<br><br> 
***Por ejemplo:***

```python
salaries[((salaries['job_title'] == 'Data Scientist') & (salaries['salary_in_usd'] == 212000)) | 
    ((salaries['job_title'] == 'Data Engineer') & (salaries['salary_in_usd'] == 300000))]['job_title'].value_counts()
```

**Otra forma de filtrar es con el metodo "isin()", el cual usaremos de la siguiente manera:** 

```python
salaries[
    (salaries['job_title'].isin(['Data Scientist','Data Engineer'])) & 
    (salaries['salary_in_usd'] > 212000)
    ]
```

```python
salaries[
    (salaries['job_title'].isin(['Data Scientist','Data Engineer'])) & 
    (salaries['salary_in_usd'] > 212000)
    ]
```

**Igual que antes, añadiremos un value_counts(), para contar los resultados**

```python
salaries[
    (salaries['job_title'].isin(['Data Scientist','Data Engineer'])) & 
    (salaries['salary_in_usd'] > 212000)
    ]['job_title'].value_counts()
```

## Operador tilde (~) para la negación bit a bit

**En pandas, ~ es el operador de NEGACIÓN lógica y equivale a decir NOT.**<br><br>

**Este se usa para identificar todos los valore que NO cumplen con los filtros presentados**<br><br>

***Por ejemplo:***

```python
salaries[~(salaries['job_title'].isin(['Data Scientist','Data Engineer']))] 
#Aqui estamos excluyendo los titulos que sean Data Engineer y Data Scientist
```

**Podemos usar este tile "~" para negar varias condicione tambien**<br><br>
***Por ejemplo:***

```python
salaries[
    (~(salaries['job_title'].isin(['Data Scientist','Data Engineer']))) &
    (~(salaries['experience_level'] == 'SE'))
    ] 
#Aqui estamos diciendo que nos busque todos aquellos registros que no sean ni Data Engineer ni Data Scientist y 
#todos los que no tengan nivel de experiencia SE
```

```python
salaries[
    ~(
        (salaries['job_title'] == 'Data Scientist') &
        (salaries['experience_level'] == 'SE')
     )
]
#Aqui por ejemplo, le estamos diciendo que ignore todos los registros que tengan titulo Data Scientist y ademas nivel SE en ese cargo
```

## Crear y Eliminar columnas

**Para crear una nueva columna, solo debemos escribir el nombre de la variable con nuestro DataFrame, seguido de corchetes y dentro de los mismos entre comillas simples el nombre de la columna, ya por ultimo, debemos agregar un igual y el valor que ocupara esa columna.**<br><br>

***Por ejemplo:***

```python
airbnb['City'] = 'New York'
```

```python
airbnb
```

**Por otro lado, para eliminar columnas usamos el metodo drop(), especificando entre parentesis la o las columna(s) a borrar**<br><br>

***Por Ejemplo:***

```python
airbnb.drop(columns = 'City', inplace = True) #Podemos guardar el resultado en una variable, o usar inplace para mantener los cambios
```

```python
airbnb
```

**Otra opcion para hacer lo mismo, es usando drop con los parametros labels y axis**<br><br>

***Ejemplo:***<br>

airbnb.drop(labels = 'City', axis=1)<br> 

**En este caso tambien podemos usar inplace para guardar los cambios en el dataframe**<br><br>

**IMPORTANTE:** <br><br>

***El valor que toma el parametro axis, dependera de si quieres borrar filas o columnas, por defecto, el metodo tiene el parametro axis en 0 (filas), el parametro se debe poner en 1 si se quiere dropear una columna.*** <br><br>

axis=0  → filas (vertical)<br>
axis=1  → columnas (horizontal)
<br><br>
**axis = 0 (default):**
- Trabaja a lo largo de las filas
- Operas columna por columna

**Se usa para:**
- Borrar filas
- Calcular estadísticas por columna

**axis = 1:**
- Trabaja a lo largo de las columnas
- Operas fila por fila

**Se usa para:**
- Borrar columnas
- Cálculos por fila
