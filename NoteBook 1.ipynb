## Cambiar nombres columnas data set

*Para cambiar los nombres de las columnas lo mas facil es usar la fucion columns, para ver el nombre de las columnas, luego copiar esos datos
en una nueva variable y cambiar en la variablers los nombres, hecho esto, usas el comando para cargar el data set, poniendo una coma seguido de
**names = [nombre de tu variable]** sin brackets, luego de esto, debes especificar la fila en la que estan los headers del csv (Usualmente la 0),
esto se hace poniendo una coma despues de la variable con los nombres, seguido de <br>
**header = Numero de columna en que estan los headers***
