### Estadistica descriptiva con metodo Describe

La funcion **describe()** no da en el mismo output la desviacion estandar, conteo de datos, promedio, entre otras estadisticas y datos de utilidad para estudiar el dataset

```python
salaries.describe() # Como podemos ver aqui nos da las estadisticas que hemos estado sacando para todas las columnas con valores numericos
```

Como podemos ver en la tabla nos da en un solo output el conteo de datos no nulos, el promedio, el valor minimo, el maximo, entre otras estadisticas y datos en las columnas work year, salary, salary in usd y remote ratio. <br>
Para los numeros que tine un +03, +05, etc, se puede escribir el numero tal cual en una celda y te dara el numero completo, por ejemplo:

```python
8.805000e+03
```

**Los porcentajes mostrados en la tabla, corresponden a los cuartiles**, los cuales son los puntos de corte del dataset, dividiendolo en 4 partes <br>
¿Por que se llaman cuartiles si son solo 3 porcentajes? Por que los porcentajes son los puntos de corte, ya que para separar algo en 4 partes necesitas 3 cortes, el 0% y el 100% tambien son percentiles, pero no cuartiles, **los cortes son 25%, 50% (Mediana) y 75%**. <br> ***Los cuartiles dividen el conjunto de datos en cuatro grupos de igual tamaño, una vez que los valores han sido ordenados de menor a mayor.***

Adicionalmente, para sacar la moda, puedes usar la funcion **mode()**, la cual **devolvera una serie de numeros (varios numeros)**, ya que puede haber mas de una moda en un conjunto de datos, por ejemplo, si en los salarios hay 2.000 personas con sueldo de 80.000 usd y 2.000 personas con sueldo de 50.000, digamos que todos los demas sueldos son distintos para este ejemplo, pero el sueldo de 80 mil y el de 50 mil son los que mas se repiten y no solo eso, se repiten la misma cantidad de veces, por lo que en este caso, hay dos modas. <br>

***Nota: si deseas sacar solo la primera moda encontrada, por ejemplo, puedes poner un indice en la funcion mode(), por ejemplo "salaries['salary_in_usd'].mode()[0]"*** 

```python
salaries['salary_in_usd'].mode()[0] # Con indice, como se ve aqui, solo me trajo la primera moda encontrada
```

```python
salaries['salary_in_usd'].mode() # sin indice, como se puede ver me devolvio las modas 0 y 150.000
```

```python
salaries.describe()
```

En este caso, podemos interpretar gracias a los valores que el **25%** de los registros, tiene un salario de 105.000 USD o menos, el **50%** 142.200 USD o menos y el **75%** tiene un salario de 185.900 USD o menos. <br>

**Por que menos?** <br> 
***Porque un cuartil no representa a un grupo, sino un punto de corte. Por ende, ese valor (sea del cuartil que sea) indica el valor maximo que alcanza un cierto porcentaje de los datos***, En el caso del primer cuartil (25%), el sueldo máximo alcanzado por el 25% inferior de los registros es de 105.000 USD

```python
salaries.describe(include='object')
```

A la funcion "describe()" se le pueden introducir distintos parametros, por ejemplo, que tipo de dato incluir. <br> 
En la celda anterior, se le introdujo el parametro "include" y se le especifico que nos incluyera los datos de tipo objeto. <br>
Como resultado, la funcion nos arrojó en "count" el conteo de los datos tipo objeto, no nulos de cada columna con ese tipo de dato, asi mismo ahora nos incluye la fila "Unique" que nos indica cuantos datos unicos hay en cada columna con valores del tipo de dato objeto. Por otro lado, tambien nos agrego la fila "top", la cual indica el valor mas frecuente de la columna, ya por ultimo, tenemos la fila "freq" la cual nos indica la frecuencia, digase, cuantas veces se repitio ese valor mas frecuente que aparece en top

Un ejemplo de interpretacion de los datos entregados en una columna por la funcion, puede ser en la columna Job_Title, en esa columna podemos ver que se encontraron 8.805 valores no nulos y 124 valores (nombres de titulos profesionales) unicos, asi mismo, podemos concluir, que el titulo mas repetido entre los valores, es el titulo de Data Engineer, el cual se repite 2.062 veces en la columna job_title

Otro parametro que se le puede dar a la funcion describe, es el parametro "percentiles", este nos permite consultar los percentiles que nosotros deseemos. por ejemplo, digamos que en lugar de los percentiles 25% 50% 75% quiero los percentiles 20%, 60% y 90%, para esto, el codigo seria con nuestro ejemplo de salarios "salaries.describe(percentiles= [0.2, 0.6, 0.9])" <br><br>

**Nota: la funcion describe, independiente de los percentiles que le demos, siempre nos va a dar de igual forma el percentil del 50%, esto ya que el 50% es la mediana de los datos**

```python
salaries.describe(percentiles=[0.2, 0.6, 0.9])
```

### Value Counts

La funcion **value_counts()** permite contar cuantos registros hay por cada valor en la columna, por ejemplo, en la siguiente celda, utilice la funcion antes mencionada para contar cuantos registros hay en la columna de job_title por cada titulo, esto con el fin de contar cuantos empleados hay por cada titulo profesional.

```python
salaries['job_title'].value_counts()
```

En este caso, podemos concluir que tenemos 2062 empleados con el titulo de Data Engineer y 908 machine learning engineers, lo que nos puede ayudar a concluir que el titulo profesional mas comun entre nuestros empleados es el de Data Engineer

La funcion **value_count()**, al igual que otras funciones de pandas, puede recibir varios parametros, los cuales son: <br>
**normalize**: permite ver proporciones y es de tipo bool <br>
**sort**: permite ordenar por frecuencia y es de tipo bool <br>
**ascending**: permite ordenar los datos en orden ascendente y es de tipo bool <br>
**bins**: permite agrupar valores numericos y puede ser de tipo int o list <br>
**dropna**: permite decidir si incluimos o no los valores NaN (vacio) y es de tipo bool

```python
# A continuacion un ejemplo del parametro ascending cuando lo ponemos en true
salaries['job_title'].value_counts(ascending=True) 
```

```python
# Otra cosa que se puede hacer con value_counts es darle varias columnas, en cuyo caso pandas calcula la frecuencia de cada combinación única 
# de valores entre dichas columnas, devolviendo una Serie con índice múltiple.
salaries[['job_title', 'salary_in_usd']].value_counts() 
```

***Con estos resultados, podemos concluir que por ejemplo, existen 68 registros donde el cargo del empleado es data engineer y el salario es exactamente 160.000 USD, lo que nos da a entender que la mayoria de empleados con el cargo de data engineer tiene un sueldo de 160.000 USD***
