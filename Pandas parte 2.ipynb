**De necesitar ordenar los datos segun los valores de otra columna, podemos utilizar la funcion "sort_values()", especificando en los parametros la columna a utilizar.** <br><br>
**Nota: por defecto siempre el orden sera de forma ascendente.**<br><br>
***Por ejemplo:*** <br>
Digamos que deseo ordenar los datos de la criptomoneda ETH con el fin de saber cual es el valor mas bajo al que llego, para esto utilizaria la funcion sort_values, especificando que usare la columna 'Price' como se ve en la celda inferior.

```python
eth.sort_values(by='Price')
```

Fijandonos en la primera fila veremos que el valor mas bajo al que llego la criptomoneda ETH, historicamente fue 107.90 dolares y si nos fijamos en la ultima fila, veremos que el ETH llego a un valor maximo de 4.808 dolares. <br><br>

**Nota: si quisieramos ordenar los datos utilizando la columna precio, pero en orden descendente, le pasariamos a sort_values() ademas el parametro de tipo bool "ascending", definiendolo en falso (Para que haga lo contrario a orden ascendente).** <br><br>

**Ejemplo:**

```python
eth.sort_values(by='Price', ascending=False)
```

**La funcion sort_values() ademas puede recibir varias columnas y diferenciar cuales columnas se solicitaron en orden ascendente y cuales en descendente.** <br><br>

**IMPORTANTE: Si utilizamos varias columnas y distintas configuraciones de orden, debemos introducir los parametros dentro de corchetes y en orden respectivo (Como se ve en el ejemplo que se presenta mas abajo).**<br><br>

**Por ejemplo:** *Digamos que ahora utilizando el data set Salaries, deseo ordenar las filas por cargo y sueldo, esto se haria de la siguiente forma:*

```python
salaries.sort_values(by= ['job_title', 'salary_in_usd'], ascending = [True, False])
```

***Para este caso, escogimos ordenar el cargo en orden descendente con el fin de verlos por orden alfabetico, pero ademas sub ordenamos los datos de sueldo en dolares en orden descendente, esto para ver primero las filas que corresponden a las personas con ese titulo que tienen el sueldo mas alto.*** <br><br>

**NOTA: Podemos ademas usar el anteriormente mencionado parametro "inplace" para guardar en la variable el DataFrame con el orden, que definimos ahora, como se ve en la celda de abajo**

```python
salaries.sort_values(by= ['job_title', 'salary_in_usd'], ascending = [True, False], inplace=True)
```

```python
salaries
```

## Accessing Data with Loc & Iloc

### **Loc[ ]**:

**La funcion loc[ ], nos permite localizar una fila de datos en particular mediante etiquetas**<br><br>

***IMPORTANTE: La funcion Loc, es con CORCHETES "[ ]" NO CON PARENTESIS "( )".***<br><br>

**Por ejemplo:**

```python
eth.loc['22/03/2020']
```

***Podemos ver los datos en este formato, o podemos agregar un par extra de corchetes para verlo en formato DataFrame.***

```python
eth.loc[['22/03/2020']]
```

**Podemos ademas consultar varios index o datos al mismo tiempo.** <br><br>

***Por ejemplo:***

```python
eth.loc[['22/03/2020', '22/03/2021','22/03/2022']]
```

**Junto con eso, podemos solicitar que nos recupere solo algunas filas, si no queremos cargarlas todas**
<br>
<br>
***Por ejemplo:***

```python
eth.loc[['22/03/2020', '22/03/2021','22/03/2022'], ['Price']]
```

```python
eth.loc[['22/03/2020', '22/03/2021','22/03/2022'], ['Price', 'Change %']]
```

**Loc puede buscar datos en cualquier columna, por lo que si por ejemplo quisieramos ver el total de registros que corresponden a un valor en la columna x, podriamos hacerlo** <br><br>

***Por ejemplo:*** *diagamos que quiero solo recuperar los registros del dataset airbnb que tengan una cama, lo haria asi:*

```python
airbnb.columns
```

```python
airbnb['Beds'].value_counts().loc[1]
```

### Iloc[ ]:

**Iloc[ ], sirve para buscar filas (registros) al igual que Loc, la diferencia y el por que tiene esa I es por que Iloc, solo recibe index numericos en formato int**

***Por ejemplo:***

```python
airbnb.iloc[[50]]
```

## Advanced indexing: Multi-Indexes ans Slices

```python
happiness #usaremnos el data set de world happiness (en el curso el pana lo llamo world)
```

```python
happiness = happiness.set_index('Country or region').sort_values(by='Country or region') #Primero cambie el index al pais
```

```python
happiness.loc['Ireland':'Canada'] 
```

*Como podemos ver aqui se intento consultar por los datos entre irlanda y canada **(los dos puntos ":" indican entre )** pero no cargo nada. Lo anterior sucedio por que como se ordena en ascendente por orden alfabetico la I va despues de la C.*

**Para obtener datos lo que haremos sera invertir el orden, poniendo primero a canada y luego a irlanda**

```python
happiness.loc['Canada':'Ireland']
```

*Como se comento antes, los dos puntos ":" significan entre, pero pandas nos permite usarlos sin dos valore, cargando asi por defecto desde el valor entregado hasta el fin del data set o desde el inicio del dataset hasta el valor entregado*

***Por ejemplo:***

```python
happiness.loc[:'Ireland'] # Esto carga desde el inicio del dataframe hasta irlanda
```

```python
happiness.loc['Canada':] #Esto Carga desde Canada hasta el fin del Dataframe
```

***Nota: para resetear el index, podemos utilizar la funcion "reset_index()"***

```python
happiness = happiness.reset_index()
```

```python
happiness
```

**Anteriormente, pudimos ver que tenemos registros de varios años para los paises en el dataset, por lo que usarl el pais como index no seria adecuado, esto debido a que la idea de un index es que sea unico, de modo que se eviten problemas por alcance de valores de index (varios registros con un mismo index)** <br><br>

**Para evitar estos problemas, podemos crear index dobles (tomar el valor de dos columnas y que desglose por ambos index)** 

***Por Ejemplo:***

```python
happiness = happiness.set_index(['Country or region', 'Year']) #Como son varias columnas hay que poner corchetes
```

```python
happiness
```

**Como podemos ver, gracias al index doble, ahora nos separa por pais y año del registro**<br><br>

*Debido a lo anterio, ahora podemos consultar por pais y el año del registro* <br><br>

***Por ejemplos:***

```python
happiness.loc['Ireland', 2019]
```

## From DataFrames to Plot (Plot en pandas)

***DATO UTIL: Los plot son los graficos***

```python
salaries
```

```python
salaries['job_title'].value_counts() 
```

***Como pudimos ver en la celda anterior, deseabamos contar la cantidad de registros por titulo de trabajo, pero digamos que queremos verlo como una tabla, en dado caso, lo que podemos hacer, es utilizar la funcion de pandas "DataFrame()", para utilizar esta funcion lo hacemos de la siguiente manera:***

```python
pd.DataFrame(salaries['job_title'].value_counts())
```

**Como pudimos ver, ya podemos ver el resultado como DataFrame, pero ahora tenemos otro problema, hay demasiadas filas, por lo que ahora me muestra solo los 5 primeros que tienen mas registros y los 5 finales que tienen menos registros, para poder cargar mas, podemos utilizar la antes aprendida funcion "head()", la cual en este caso usaremos para cargar los primeros 20 titulos con mas repeticion, lo que se hace de la siguiente forma:**

```python
top20jobs = pd.DataFrame(salaries['job_title'].value_counts()).head(20)
```

```python
top20jobs
```

***Nota: Podemos usar la funcion tail() de la misma forma, pero esta nos cargara los ultimos n valores***

Digamos que ahora deseo graficar esta tabla, eso lo haremos con la funcion "plot()"

```python
top20jobs.plot()
```

***Como podemos ver, este grafico, no nos es muy util, pero podemos cambiar su tipo si le entregamos el parametro "kind"***
<br><br>
**Los tipos pueden ser:** <br><br>
‘line’ : line plot (default)

‘bar’ : vertical bar plot

‘barh’ : horizontal bar plot

‘hist’ : histogram

‘box’ : boxplot

‘kde’ : Kernel Density Estimation plot

‘density’ : same as ‘kde’

‘area’ : area plot

‘pie’ : pie plot

‘scatter’ : scatter plot (DataFrame only)

‘hexbin’ : hexbin plot (DataFrame only)
<br><br>
**Para este caso usaremos el tipo bar, que nos da un grafico de barras**

```python
top20jobs.plot(kind='bar')
```

Como podemos ver, este grafico ya es mucho mas util, pero hagamos un grafico horizontal para probar otros tipos de grafico tambien

```python
top20jobs.plot(kind='barh')
```

**Nota: si quisieramos poner los valores con mas repeticion arriba (lo que personalmente prefiero), podemos ordenarlo utilizando la funcion "sort_values()", de la siguiente forma:**

```python
top20jobs.sort_values(by= 'count').plot(kind='barh')
```

***IMPORTANTE: La funcion sort_values, va primero que plot, esto debido a que sort_values prepara los datos y plot solo dibuja lo que recibe.*** <br><br>

**De hacerlo al reves, plot(), devolvera un objeto Axes y los objetos Axes, no tienen sort_values()**

Otro parametro que le podemos dar a plot, es **color**, que tal cual dice su nombre, nos permite cambiar el color del grafico. <br><br>

***Por ejemplo:***

```python
top20jobs.sort_values(by= 'count').plot(kind='barh', color='pink')
```
